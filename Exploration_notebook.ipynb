## TITLE 6. Cash Inflow Analysis

Cash inflow categories were extracted and analyzed to identify the main revenue sources of the company during the reporting period.

```python
import pandas as pd

cash_in = pd.read_excel(
    'поступление денежных средств.xlsx',
    header=None
)

cash_in = cash_in.iloc[36:].reset_index(drop=True)

display(cash_in.head(15))
```

##   TITLE 7. Cash Outflow Analysis


Cash outflow categories were analyzed to determine the largest expense groups and financial obligations.

```python
import pandas as pd

cash_out = pd.read_excel(
    'расход денежных средств.xlsx',
    header=None
)

cash_out = cash_out.iloc[36:].reset_index(drop=True)

display(cash_out.head(15))
```

## TITLE 8. Top Cash Inflow Sources

The largest cash inflow categories were visualized to identify the primary revenue sources of the company.

```python
import pandas as pd
import matplotlib.pyplot as plt

inflow = cash_in.iloc[2:6].copy()

inflow = inflow[[0,13]]
inflow.columns = ['Category','Total']

inflow['Total'] = pd.to_numeric(
    inflow['Total'],
    errors='coerce'
)

plt.figure(figsize=(8,5))
plt.bar(inflow['Category'], inflow['Total'])
plt.title('Top Cash Inflow Sources')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

# TITLE 3. Detailed Dataset Inspection

Description

At this stage, each ERP dataset is inspected to identify useful business indicators, reporting structures, and analytical opportunities.

```python
import pandas as pd

files = [
    'анализ счета бух.xlsx',
    'движение товаров.xlsx',
    'материальная ведомость.xlsx',
    'номенклатура.xlsx',
    'оборачиваемость товаров.xlsx',
    'оборотно-сальдовая ведомость.xlsx',
    'оборотные средства.xlsx',
    'поступление денежных средств.xlsx',
    'расход денежных средств.xlsx'
]

for file in files:
    print("\n" + "="*80)
    print(file)

    try:
        df = pd.read_excel(file, header=None)

        print("Rows:", df.shape[0])
        print("Columns:", df.shape[1])

        for i in range(45):
            row = df.iloc[i].tolist()

            if any(str(x) != 'nan' for x in row):
                print(f"ROW {i}:")
                print(row)

    except Exception as e:
        print(e)
```
